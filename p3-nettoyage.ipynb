```python
data = data.reset_index(drop=True)

data.drop(['Comments', 'ZipCode'], axis=1, inplace=True)
```

Ici on remplace les valeurs NaN par 0 pour éviter les erreurs lors de la prédiction

```python
columns_abberante = ['PropertyGFABuilding(s)','SourceEUI(kBtu/sf)','SourceEUIWN(kBtu/sf)', 'PropertyGFAParking','Electricity(kWh)','Electricity(kBtu)','TotalGHGEmissions','GHGEmissionsIntensity']

for col in columns_abberante:
  data[col] = data[col].apply(lambda x : x if x >= 0 else 0)
```

On note toute les colonnes qui ont des valeurs non supérieur ou égal à 0 pour les remplacer ensuite par 0 

Maintenant, on va check nos data pour voir si les changements ont bien était fais

```python
data.describe()
```

On remarque des variables qui sont encore en NaN 

On va mettre leur valeur à la moyenne pour avoir des données plus ou moins réaliste

```python
data[['ThirdLargestPropertyUseType', 'SecondLargestPropertyUseType']] = data[['ThirdLargestPropertyUseType', 'SecondLargestPropertyUseType']].fillna('No Use')
data['LargestPropertyUseTypeGFA'] = data['LargestPropertyUseTypeGFA'].fillna(np.mean(data['LargestPropertyUseTypeGFA']))
data['LargestPropertyUseType'] = data['LargestPropertyUseType'].fillna('Unknown')
data[['ThirdLargestPropertyUseTypeGFA', 'SecondLargestPropertyUseTypeGFA']] = data[['ThirdLargestPropertyUseTypeGFA', 'SecondLargestPropertyUseTypeGFA']].fillna(np.mean(data[['ThirdLargestPropertyUseTypeGFA', 'SecondLargestPropertyUseTypeGFA']]))

data.isna().sum().sort_values()
```

On va refaire pareil pour les autres valeurs

```python
fuel_data = ['SiteEnergyUse(kBtu)','SiteEUIWN(kBtu/sf)','SiteEnergyUseWN(kBtu)','SiteEUI(kBtu/sf)','NaturalGas(kBtu)','NaturalGas(therms)','SteamUse(kBtu)']
for item in fuel_data:
  data[item] = data[item].fillna(np.mean(data[item])) 

data.isna().sum().sort_values()
```

On remarque encore des valeurs NaN pour certaines colonnes

On va donc continuer le nettoyage

```python
data.dropna(subset=['ENERGYSTARScore','TaxParcelIdentificationNumber'],inplace=True) 
data['NumberofFloors'] = data['NumberofFloors'].fillna(np.mean(data['NumberofFloors']))

lbs_outlier = {
  'High outlier':'High Outlier',
  'Low outlier':'Low Outlier',
   np.nan:'NAN'
}

data['Outlier'].replace(lbs_outlier, inplace=True)

data = data[data['Outlier'] != 'High Outlier']

drop_col = ['Outlier', 'ListOfAllPropertyUseTypes','DefaultData','YearsENERGYSTARCertified']
data.drop(columns=drop_col,inplace=True) 

data.isna().sum().sum() 
```

Ici on retire les colonnes inutile, egarde que les valeurs outlier qui ne sont pas haute et on met les autres valeurs à la moyenne puis on vérifie le nombre de NaN
